```python
import pandas as pd

file_path = "/content/Sentiment Analysis Dataset.csv"  # Update with the correct path if needed

# Use 'on_bad_lines' instead of 'error_bad_lines'
# 'skip' will skip bad lines
df = pd.read_csv(file_path, delimiter=",", quoting=3, encoding="utf-8", on_bad_lines='skip')

# Check dataset structure
print(df.info())
print(df.head())
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab') # Download the punkt_tab data

# Initialize tools
stop_words = set(stopwords.words("english"))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Text preprocessing function
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    tokens = word_tokenize(text)  # Tokenization
    tokens = [word for word in tokens if word.isalnum()]  # Remove punctuation
    tokens = [word for word in tokens if word not in stop_words]  # Remove stopwords
    stemmed = [stemmer.stem(word) for word in tokens]  # Stemming
    lemmatized = [lemmatizer.lemmatize(word) for word in stemmed]  # Lemmatization
    return " ".join(lemmatized)

# Apply preprocessing to the text column (update column name accordingly)
df["ProcessedText"] = df["SentimentText"].astype(str).apply(preprocess_text)

# Display processed data
print(df[["SentimentText", "ProcessedText"]].head())
```

```python

# Core dependencies
from google.colab import files  # For file uploads in Colab environment
import pandas as pd  # Data manipulation and analysis
import numpy as np  # Numerical computing
from sklearn.model_selection import train_test_split, cross_val_score  # Data splitting and validation
from sklearn.feature_extraction.text import TfidfVectorizer  # Text feature extraction
from sklearn.linear_model import LogisticRegression  # Classification model
from sklearn.metrics import classification_report, confusion_matrix  # Model evaluation
import matplotlib.pyplot as plt  # Visualization
import seaborn as sns  # Enhanced visualization
import joblib  # Model serialization

# =====================================================================
# DATA LOADING & PREPROCESSING
# =====================================================================

# Robust CSV reading with multiple fallback strategies
try:
    # Primary attempt with UTF-8 encoding and modern error handling
    df = pd.read_csv('/content/Sentiment Analysis Dataset.csv',
                    on_bad_lines='warn',  # Skip malformed rows with warning
                    encoding='utf-8')  # Standard text encoding
except Exception as e:
    print(f"Primary read failed: {e}")
    try:
        # Fallback to Python engine with Latin-1 encoding (broader character support)
        df = pd.read_csv('/content/Sentiment Analysis Dataset.csv',
                        engine='python',  # More flexible parser
                        encoding='latin1')  # Wider character encoding
    except Exception as e:
        print(f"Secondary read failed: {e}")
        raise SystemExit("Fatal error: Could not load dataset")

# =====================================================================
# EXPLORATORY DATA ANALYSIS (EDA)
# =====================================================================

print(f"\n{' DATASET OVERVIEW ':=^50}")
print(f"Shape: {df.shape} (rows, columns)")
print(f"\n{' SAMPLE RECORDS ':-^50}")
print(df.head())  # Show structural overview

# Data quality assessment
print(f"\n{' DATA QUALITY CHECK ':-^50}")
missing_values = df.isnull().sum()
print("Missing values per column:")
print(missing_values[missing_values > 0])  # Only show columns with missing data

# Class distribution analysis
print(f"\n{' CLASS DISTRIBUTION ':-^50}")
class_dist = df['Sentiment'].value_counts()
print(class_dist)

# Visualize class balance
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.countplot(x='Sentiment', data=df)
plt.title('Class Distribution\n(0=Negative, 1=Positive)')
plt.xlabel('Sentiment Label')
plt.ylabel('Count')

# =====================================================================
# FEATURE ENGINEERING
# =====================================================================

# Stratified train-test split (preserves class distribution)
X = df['SentimentText']  # Text features
y = df['Sentiment']  # Target labels
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,  # Industry-standard 80/20 split
    random_state=42,  # Reproducibility seed
    stratify=y  # Maintain original class ratios
)

# Advanced text vectorization with TF-IDF
vectorizer = TfidfVectorizer(
    max_features=5000,  # Dimensionality control
    stop_words='english',  # Remove common words
    ngram_range=(1, 2),  # Capture unigrams and bigrams
    min_df=5,  # Ignore rare terms (<5 docs)
    max_df=0.7  # Ignore overly common terms (>70% docs)
)
X_train_vec = vectorizer.fit_transform(X_train)  # Fit on training data only
X_test_vec = vectorizer.transform(X_test)  # Transform test data

# =====================================================================
# MODEL TRAINING & VALIDATION
# =====================================================================

# Configure logistic regression with class balancing
model = LogisticRegression(
    max_iter=1000,  # Sufficient iterations for convergence
    class_weight='balanced',  # Handle imbalanced classes
    random_state=42,  # Reproducibility
    solver='liblinear'  # Optimal for medium-sized datasets
)

# Cross-validated performance estimation
print(f"\n{' CROSS-VALIDATION RESULTS ':=^50}")
cv_scores = cross_val_score(
    model, X_train_vec, y_train,
    cv=5,  # 5-fold validation
    scoring='accuracy'  # Evaluation metric
)
print(f"Fold Accuracies: {cv_scores}")
print(f"Mean Accuracy: {np.mean(cv_scores):.3f} ± {np.std(cv_scores):.3f} (SD)")

# Final model training
model.fit(X_train_vec, y_train)
print("\nModel training completed")

# =====================================================================
# MODEL EVALUATION
# =====================================================================

# Generate predictions and comprehensive report
y_pred = model.predict(X_test_vec)
print(f"\n{' CLASSIFICATION REPORT ':=^50}")
print(classification_report(y_test, y_pred,
                          target_names=['Negative', 'Positive'],
                          digits=4))  # 4 decimal places

# Confusion matrix visualization
cm = confusion_matrix(y_test, y_pred)
plt.subplot(1, 2, 2)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Negative', 'Positive'],
            yticklabels=['Negative', 'Positive'])
plt.title('Confusion Matrix\n(Counts)')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.tight_layout()
plt.show()

# =====================================================================
# MODEL INTERPRETATION
# =====================================================================

# Extract and visualize most predictive features
feature_names = vectorizer.get_feature_names_out()
coef = model.coef_[0]
top_df = pd.DataFrame({'Word': feature_names,
                      'Coefficient': coef}).sort_values('Coefficient')

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
top_df.tail(10).plot.barh(x='Word', y='Coefficient', color='green')
plt.title('Top 10 Positive Indicators')
plt.xlabel('Coefficient Magnitude')

plt.subplot(1, 2, 2)
top_df.head(10).plot.barh(x='Word', y='Coefficient', color='red')
plt.title('Top 10 Negative Indicators')
plt.tight_layout()
plt.show()

# =====================================================================
# OPERATIONALIZATION
# =====================================================================

# Example predictions for model validation
test_phrases = [
    "I love this product!",  # Expected positive
    "This is terrible quality",  # Expected negative
    "The service was adequate"  # Neutral/ambiguous
]
test_vec = vectorizer.transform(test_phrases)
predictions = model.predict(test_vec)

print(f"\n{' EXAMPLE PREDICTIONS ':=^50}")
for text, pred in zip(test_phrases, predictions):
    print(f"Text: {text[:50]}... → Prediction: {'Positive' if pred == 1 else 'Negative'}")

# Model persistence for production use
joblib.dump(model, 'sentiment_model.pkl')
joblib.dump(vectorizer, 'vectorizer.pkl')
print(f"\n{' MODEL PERSISTENCE ':-^50}")
print("Model artifacts saved to disk:")
print("- sentiment_model.pkl (trained classifier)")
print("- vectorizer.pkl (feature engineering pipeline)")
```

**Model Performace**

```python
# =====================================================================
# SAVED MODEL ANALYSIS PIPELINE
# =====================================================================
"""
COMPREHENSIVE MODEL ANALYSIS
---------------------------
This script analyzes a persisted sentiment analysis model by:
1. Loading the serialized model and vectorizer
2. Recreating the evaluation environment
3. Performing feature importance analysis
4. Conducting prediction validation
5. Generating interpretability reports
"""

import joblib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix

# =====================================================================
# 1. MODEL LOADING AND VALIDATION
# =====================================================================

# Load persisted artifacts
print("Loading model artifacts...")
try:
    model = joblib.load('sentiment_model.pkl')
    vectorizer = joblib.load('vectorizer.pkl')
    print("✅ Model and vectorizer loaded successfully")
except Exception as e:
    raise SystemExit(f"❌ Error loading model: {e}")

# Verify model components
print("\nModel Class:", type(model).__name__)
print("Model Parameters:")
print(model.get_params())

# =====================================================================
# 2. FEATURE ANALYSIS
# =====================================================================

# Extract model coefficients and feature names
if hasattr(model, 'coef_'):
    feature_names = vectorizer.get_feature_names_out()
    coefficients = model.coef_[0]

    # Create feature importance dataframe
    features_df = pd.DataFrame({
        'Feature': feature_names,
        'Coefficient': coefficients,
        'Absolute_Impact': np.abs(coefficients)
    }).sort_values('Absolute_Impact', ascending=False)

    # Top 20 most influential features
    print("\nTop 20 Predictive Features:")
    display(features_df.head(20).style\
        .background_gradient(cmap='RdBu', subset=['Coefficient'])\
        .format({'Coefficient': '{:.4f}'}))

    # Visualize feature impacts
    plt.figure(figsize=(12, 8))
    top_n = 15
    top_features = features_df.head(top_n)

    plt.subplot(2, 1, 1)
    sns.barplot(x='Coefficient', y='Feature',
                data=top_features, palette='viridis')
    plt.title(f'Top {top_n} Predictive Features')
    plt.xlabel('Coefficient Magnitude')
    plt.ylabel('')

    plt.subplot(2, 1, 2)
    sns.histplot(features_df['Coefficient'], bins=50, kde=True)
    plt.title('Distribution of Feature Coefficients')
    plt.xlabel('Coefficient Value')
    plt.ylabel('Count')

    plt.tight_layout()
    plt.show()

# =====================================================================
# 3. PREDICTION ANALYSIS
# =====================================================================

# Create test cases for prediction validation
test_cases = [
    ("This product works great and I love it!", "Expected: Positive"),
    ("Terrible experience would not recommend", "Expected: Negative"),
    ("It's okay but could be better", "Expected: Neutral"),
    ("The service was acceptable", "Expected: Neutral"),
    ("Worst purchase ever complete waste of money", "Expected: Negative")
]

print("\nPrediction Validation:")
results = []
for text, expectation in test_cases:
    vec = vectorizer.transform([text])
    proba = model.predict_proba(vec)[0]
    pred = model.predict(vec)[0]
    results.append({
        'Text': text[:40] + '...' if len(text) > 40 else text,
        'Expectation': expectation,
        'Prediction': 'Positive' if pred == 1 else 'Negative',
        'P(Negative)': f"{proba[0]:.1%}",
        'P(Positive)': f"{proba[1]:.1%}"
    })

display(pd.DataFrame(results))

# =====================================================================
# 4. MODEL INTERPRETABILITY
# =====================================================================

def explain_prediction(text, top_n=5):
    """Explain model predictions by showing top influential features"""
    vec = vectorizer.transform([text])
    features = vectorizer.get_feature_names_out()
    coefs = model.coef_[0]
    feature_values = vec.toarray()[0]

    # Get influential features
    influential = pd.DataFrame({
        'Feature': features,
        'Value': feature_values,
        'Coefficient': coefs,
        'Impact': feature_values * coefs
    }).sort_values('Impact', key=abs, ascending=False)

    # Display explanation
    print(f"\nText: '{text}'")
    print("Top Influential Features:")
    display(influential.head(top_n).style\
        .bar(color=['#ff7f7f', '#7fbf7f'], align='zero', subset=['Impact'])\
        .format({'Impact': '{:.4f}', 'Coefficient': '{:.4f}'}))

    proba = model.predict_proba(vec)[0]
    print(f"\nFinal Prediction: {'Positive' if proba[1] > 0.5 else 'Negative'} "
          f"(P={max(proba):.1%})")

# Example explanations
print("\nModel Explanation Examples:")
explain_prediction("The battery life is amazing but the camera is terrible")
explain_prediction("Not worth the high price in my opinion")

# =====================================================================
# 5. DECISION BOUNDARY ANALYSIS
# =====================================================================

if hasattr(model, 'predict_proba'):
    # Generate synthetic confidence scores
    test_samples = [
        "love happy great excellent fantastic",
        "hate bad terrible awful horrible",
        "average okay medium normal regular"
    ]

    print("\nDecision Boundary Analysis:")
    for text in test_samples:
        vec = vectorizer.transform([text])
        proba = model.predict_proba(vec)[0]
        print(f"\nText: '{text}'")
        print(f"P(Negative): {proba[0]:.1%}")
        print(f"P(Positive): {proba[1]:.1%}")
        print(f"Classification Threshold: {model.predict(vec)[0]}")

# =====================================================================
# 6. MODEL PERSISTENCE VERIFICATION
# =====================================================================

print("\nModel Persistence Verification:")
print(f"Model Type: {type(model).__name__}")
print(f"Vectorizer Type: {type(vectorizer).__name__}")
print(f"Vocabulary Size: {len(vectorizer.vocabulary_)}")
print("Model successfully loaded and analyzed")
```
